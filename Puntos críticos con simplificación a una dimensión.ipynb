```python
# Importamos las paqueterías que necesitamos
import matplotlib.pyplot as plt
import numpy as np
import sympy

#Declaramos los parámetros
S1 = 1.5
S2 = 1
S3 = 1
S4 = 1
K1 = 720
K2 = 720
K3 = 0.007
K4 = 0.007
daSYN = 15
kaSYN = 8.5
# Definimos a v1 (generación de ROS) y v2 (degradación de ROS)
aSYN = np.linspace(0,22,2000)
v1 = K1*(1+S1+daSYN*(((aSYN/(kaSYN))**4)/(1+(aSYN/(kaSYN))**4)))/10**3
v2 = ((K2*K4*aSYN*S2*S4)/(K3*S3))/10**3

#Graficamos
#v1
plt.plot(aSYN,v1,label='v1')
#v2
plt.plot(aSYN,v2,label='v2')
#Puntos de equilibrio (equivalen a v1-v2)
idx = np.argwhere(np.diff(np.sign(v1 - v2))).flatten()
#Los graficamos de rojo
plt.plot(aSYN[idx], v1[idx], 'ro')

#Título de la gráfica
plt.title('Sistema en una dimensión y puntos de equilibrio con S1=1.5')

#Títulos de los ejes
plt.xlabel('aSYN (microM)')
plt.ylabel('v1,v2 (miliM/h)')

#Añadimos una cuadrícula
plt.grid(alpha=.4,linestyle='--')

#Graficamos nuestras leyendas
plt.legend()

#Mostramos la gráfica
plt.show()
```
